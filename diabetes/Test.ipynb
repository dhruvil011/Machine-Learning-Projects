```python
col_names=["pregnancies","glucose","BP","skin","insulin","BMI","DiabetesPedigree","age","output"]
```

```python
new_df=pd.read_csv('diabetes.csv',names=col_names)
new_df
```

```python
new_df.columns
```

```python
new_df=new_df.iloc[1:]
new_df
```

```python
cols=["pregnancies","glucose","BP","insulin","BMI","DiabetesPedigree","age"]
x=new_df[cols]
y=new_df.output
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.25, random_state=0)
print(x_train.shape,x_test.shape)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
y_pred=logreg.predict(x_test)
y_pred
```

```python
y_test
```

```python
for i,j in zip(y_test,y_pred):
    print(i,j)
```

```python
from sklearn import metrics
```

```python
print("Accurracy",metrics.accuracy_score(y_test,y_pred))
```
